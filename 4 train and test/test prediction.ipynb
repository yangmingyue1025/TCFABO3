# 数据导入

```python
import os
import numpy as np
import pandas as pd
import math
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
from glob import glob
import warnings 
warnings.filterwarnings("ignore")
```

# 预测结果计算

```python
input_path = '/bagging/'
data = pd.read_csv(f"{input_path}data.csv")
r2_test = pd.read_csv(f"{input_path} r2_test.csv")
r2_train = pd.read_csv(f"{input_path} r2_train.csv")
rmse_test = pd.read_csv(f"{input_path} rmse_test.csv")
rmse_train = pd.read_csv(f"{input_path} rmse_train.csv")
pred_test = pd.read_csv(f"{input_path} pred_test.csv")
pred_train = pd.read_csv(f"{input_path} pred_train.csv")
y = data.iloc[:,-1] 
test_data = pred_test.iloc[:,1:]
train_data = pred_train.iloc[:,1:]
```

```python
L = len(y)
Y_test_mean = test_data.mean(axis = 1)
Y_train_mean = train_data.mean(axis = 1)
Y_test_std = test_data.std(axis = 1)
Y_train_std = train_data.std(axis = 1)
#AE
AE_test =  pd.DataFrame()
AE_train =  pd.DataFrame()
for i in range(0,L):
    name = f'K{i}'
    AE_test[name] = test_data.iloc[i,:] - y[i]
AE_test = AE_test.T
AE_test_mean = AE_test.mean(axis = 1)
AE_test_mean = AE_test_mean.to_frame()
AE_test_mean.reset_index(inplace = True)
for i in range(0,L):
    name = f'K{i}'
    AE_train[name] = train_data.iloc[i,:] - y[i]
AE_train = AE_train.T
AE_train_mean = AE_train.mean(axis = 1)
AE_train_mean = AE_train_mean.to_frame()
AE_train_mean.reset_index(inplace = True)
#\σ\
xigema_test = pd.DataFrame()
xigema_train = pd.DataFrame()
for i in range(0,L):
    name = f'K{i}'
    xigema_test[name] = (abs(test_data.iloc[i,:] - y[i]))/y[i]
xigema_test = xigema_test.T
xigema_test_mean = xigema_test.mean(axis = 1)
xigema_test_mean = xigema_test_mean.to_frame()
xigema_test_mean.reset_index(inplace = True)
for i in range(0,L):
    name = f'K{i}'
    xigema_train[name] = (abs(train_data.iloc[i,:] - y[i]))/y[i]
xigema_train = xigema_train.T
xigema_train_mean = xigema_train.mean(axis = 1)
xigema_train_mean = xigema_train_mean.to_frame()
xigema_train_mean.reset_index(inplace = True)
```

```python
result = pd.DataFrame()
result['TCF'] = y
result['train_mean'] = Y_train_mean
result['test_mean'] = Y_test_mean
result['train_std'] = Y_train_std
result['test_std'] = Y_test_std
result['train_xigema'] = xigema_train_mean[0]
result['test_xigema'] = xigema_test_mean[0]
result['train_AE'] = AE_train_mean[0]
result['test_AE'] = AE_test_mean[0]
```

```python
xigema_mean_train = (y - Y_train_mean)**2
x_train = xigema_mean_train.to_frame()
xigema_mean_test = (y - Y_test_mean)**2
x_test = xigema_mean_test.to_frame()
result['r2_train'] = r2_train.mean(axis = 1)
result['r2_test'] = r2_test.mean(axis = 1)
result['rmse_train'] = rmse_train.mean(axis = 1)
result['rmse_test'] = rmse_test.mean(axis = 1)
result['xigema_mean_train'] = math.sqrt(x_train.mean(axis = 0))
result['xigema_mean_test'] = math.sqrt(x_test.mean(axis = 0))
```
