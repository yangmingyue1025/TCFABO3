```python
dizhi = '/your folder/'
```

# data

```python
IncNodePurity = pd.read_csv(f'{dizhi}rf_im1.csv')
IncMSE = pd.read_csv(f'{dizhi}rf_im2.csv')
Rel_Inf = pd.read_csv(f'{dizhi}1-gbdt_im1.csv')
xgb_importance = pd.read_csv(f'{dizhi}xgb_im1.csv')
#svr_importance = pd.read_csv(f'{dizhi}svr_im1.csv')
name = pd.read_csv(f'{dizhi}ABO4135.csv')
name = name.iloc[:,:-1]
IncNodePurity = IncNodePurity.iloc[:,1:]
IncMSE = IncMSE.iloc[:,1:]
Rel_Inf = Rel_Inf.iloc[:,1:]
```

```python
name
```

# IncNodePurity

```python
row_mean_1 = IncNodePurity.mean(axis=1)
row_mean_1 = (row_mean_1 - row_mean_1.min()) / (row_mean_1.max() - row_mean_1.min())
IncNodePurity_result = pd.DataFrame(row_mean_1)
IncNodePurity_result.columns = ['IncNodePurity']
IncNodePurity_result['Feature'] = name.columns
IncNodePurity_result.sort_values(by = 'IncNodePurity', ascending = False)
```

# IncMSE

```python
row_mean_2 = IncMSE.mean(axis=1)
row_mean_2 = (row_mean_2 - row_mean_2.min()) / (row_mean_2.max() - row_mean_2.min())
IncMSE_result = pd.DataFrame(row_mean_2)
IncMSE_result.columns = ['%IncMSE']
IncMSE_result['Feature'] = name.columns
IncMSE_result.sort_values(by = '%IncMSE', ascending = False)
```

# Rel_Inf

```python
row_mean_3 = Rel_Inf.mean(axis=1)
row_mean_3 = (row_mean_3 - row_mean_3.min()) / (row_mean_3.max() - row_mean_3.min())
Rel_Inf_result = pd.DataFrame(row_mean_3)
Rel_Inf_result.columns = ['Rel_Inf']
Rel_Inf_result['Feature'] = name.columns
Rel_Inf_result.sort_values(by = 'Rel_Inf', ascending = False)
```

# xgb_importance

```python
xgb = xgb_importance.iloc[:,2:]
feature = xgb_importance.iloc[:,1]
Feature = pd.DataFrame(feature)
row_mean_4 = xgb.mean(axis = 1)
row_mean_4 = (row_mean_4-row_mean_4.min())/(row_mean_4.max()-row_mean_4.min())
xgb_im = pd.DataFrame(row_mean_4)
xgb_im.columns = ['xgb_importance']
xgb_im['Feature'] = Feature['Feature']
```

# save data

```python
IncNodePurity_result.to_csv(f'{dizhi}IncNodePurity.csv', index = False)
IncMSE_result.to_csv(f'{dizhi}IncMSE.csv', index = False)
Rel_Inf_result.to_csv(f'{dizhi}Rel_Inf.csv', index = False)
xgb_im.to_csv(f'{dizhi}xgb_importance.csv', index = False)
#svr_rbf.to_csv(f'{dizhi}svr.rbf_importance.csv', index = False)
```

# importance

```python
IncNodePurity_result = pd.read_csv(f'{dizhi}IncNodePurity.csv')
IncMSE_result = pd.read_csv(f'{dizhi}IncMSE.csv')
Rel_Inf = pd.read_csv(f'{dizhi}Rel_Inf.csv')
xgb_importance = pd.read_csv(f'{dizhi}xgb_importance.csv')
#svr_rbf_importance = pd.read_csv(f'{dizhi}svr.rbf_importance.csv')
Feature =  IncNodePurity_result.merge(IncMSE_result, on = ['Feature'], how = 'left')
Feature = Feature.merge(Rel_Inf, on = ['Feature'], how = 'left')
Feature = Feature.merge(xgb_importance, on = ['Feature'], how = 'left')
#Feature = Feature.merge(svr_rbf_importance, on = ['Feature'], how = 'left')
Feature['total'] = Feature['IncNodePurity'] + Feature['%IncMSE'] + Feature['Rel_Inf'] + Feature['xgb_importance'] 
#Feature['total'] = Feature['IncNodePurity'] + Feature['%IncMSE'] + Feature['Rel_Inf'] + Feature['xgb_importance'] + Feature['svr.rbf_importance']
max_1 = Feature['total'].max(axis=0)
Feature['IncNodePurity_p'] = Feature['IncNodePurity']/max_1
Feature['%IncMSE_p'] = Feature['%IncMSE']/max_1
Feature['Rel_Inf_p'] = Feature['Rel_Inf']/max_1
Feature['xgb_importance_p'] = Feature['xgb_importance']/max_1
#Feature['svr.rbf_importance_p'] = Feature['svr.rbf_importance']/max_1
Feature = Feature.sort_values(by = 'total', ascending = True)
Feature.to_csv(f'{dizhi}importance.csv', index=False)
```
