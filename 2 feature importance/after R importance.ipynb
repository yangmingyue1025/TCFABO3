## xgb

```python
dizhi = '/your folder/'
datas = glob(f'{dizhi}xgb/*.csv')
```

```python
datas
```

```python
r2_te = pd.DataFrame()

for k in range(100,310,50):
    r2_te_sum = pd.DataFrame(0 , index = [0, 1, 2, 3], 
                         columns = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10'])
    for j in range(1, 31):
        name = f' {j} {k} r2_test.csv'
        for data in datas:
            if os.path.basename(data) == name:
                file = pd.read_csv(data)
                file = file.iloc[:,1:] 
                r2_te_sum = r2_te_sum + file
    r2_te_sum = r2_te_sum/30 
    r2_te_sum.index = [1, 2, 3, 4]
    r2_te_sum.columns = [0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50]
    r2_te_sum_find = r2_te_sum.max()
    fc_max_r2_te = r2_te_sum_find.max()
    for index, row in r2_te_sum.iterrows():
        for column in r2_te_sum.columns:
            value = row[column]
            if value == fc_max_r2_te:
                r2_te = r2_te.append({ 'r2_te':fc_max_r2_te, 'nround':k, 'max_depth':index, 'eta':column}, ignore_index = True)
```

```python
r2_te
```

```python
r2_te.to_csv(f'{dizhi}xgb（nround100-300，max_depth1-4，eta0.05-0.5）.csv',index = False)
```
