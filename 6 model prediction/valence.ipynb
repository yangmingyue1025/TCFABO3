# 导入库

```python
import pandas as pd
import os
import glob
import json


valence_csv_path = 'valence.csv'  
coordination_folder_path = '/coordination number change cubic/'  
output_folder = '/coordination number valence change cubic/' 


if not os.path.exists(output_folder):
    os.makedirs(output_folder)


valence_df = pd.read_csv(valence_csv_path)


for coordination_file in glob.glob(os.path.join(coordination_folder_path, '*.csv')):
    coordination_name = os.path.basename(coordination_file)
    coordination_df = pd.read_csv(coordination_file)


    for index, row in valence_df.iterrows():
        mp_id = row['mp_id']
        formula = row['formula']
        element_valence_str = row['element_valence']


        if pd.isnull(element_valence_str) or element_valence_str.strip() == '':
            print(f"Skipping {mp_id}: element_valence is empty or invalid.")
            continue

        try:

            element_valence = json.loads(element_valence_str.replace("'", '"'))
        except json.JSONDecodeError as e:
            print(f"Skipping {mp_id}: JSONDecodeError - {e}")
            continue


        if mp_id in coordination_name:

            coordination_df['V'] = coordination_df['ion'].map(element_valence)


            output_file = os.path.join(output_folder, coordination_name)
            coordination_df.to_csv(output_file, index=False)
            print(f"Processed {coordination_name} and saved to {output_file}")

print("Processing complete.")
```
