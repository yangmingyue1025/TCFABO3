# CSV

```python
feature_combination = pd.DataFrame()
features = pd.DataFrame(data.columns)
features.columns = ['feature']
features = features.iloc[:-1,:] 
for i in range(1,len(features)+1):
    num_features = i
    combinations = list(itertools.combinations(features['feature'], num_features))
    for c in combinations:
        s = '+'.join(c)
        feature_combination = feature_combination.append({'feature':s}, ignore_index=True)
```

```python
for i in feature_combination.values.tolist():
    result = pd.DataFrame()
    after_split = i[0].split("+")
    for j in after_split:
        result[j] = data[j]
    result['TCF'] = data['TCF']
    name = 'TCF'+'~'+i[0]
    result.to_csv(f'{dizhi}特征组合文件/{name}.csv', index=False)
```
